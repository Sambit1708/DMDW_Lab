**1) Read roll numbers and marks of 5 students as two list nextperform bar plot over the data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
roll = []
marks = []

for i in range(5):
  a = input()
  roll.append(a)
  b = int(input())
  marks.append(b)
```

```python
plt.bar(roll, marks, color='blue')
plt.xlabel("Roll Number")
plt.ylabel("Marks")
plt.title("Students marks over different roll no.")
plt.show()
```

**3) Let's assume pre = np.array(19, 6, 11, 9) and post = np.array(15, 11, 9, 8) are the two numpy array, plot this two array as side by side bar plot.**

```python
pre = np.array([19, 6, 11, 9])
post = np.array([5, 11, 9, 8])

show = ['A', 'B', 'C', 'D']
X_axis = np.arange(len(show))
  
plt.bar(X_axis - 0.2, pre, 0.4, label = 'Pre')
plt.bar(X_axis + 0.2, post, 0.4, label = 'Post')

plt.xticks(X_axis, show)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
